```python
#immutable nature of tuples
n=[1,2,3,4]
n[0]=5
print(n)
m=(1,1,1,1,1)
#m[0]=5   #TypeError: 'tuple' object does not support item assignment
print(m)
```

```python
#Tuple methods
print(m.count(1))
print(m.index(1))
```

```python
#PACKING AND UNPACKING IN TUPLE
packing_tuple=1,2,3
print(packing_tuple)
a,b,c=packing_tuple  #UNPACKING
print(a)
print(b)
print(c)
#UNPACKING WITH *
number=(1,2,3,4,5,6,7,8,9)
d,*e,f=number
print(d)
print(e)
print(f)

```
