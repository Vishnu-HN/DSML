```python
"""Write a program that calculates the factorial 
of a number input by the user using a while loop."""
num = int(input("Enter the number : "))
fact =1
while(num>0):
    fact=fact*num
    num-=1

print(fact)
```

```python
"""Write a program that calculates the sum of the digits of a number 
input by the user using a while loop."""
num = int(input("Enter the number : "))
sum=0
while(num>0):
    rem=num%10
    sum +=rem
    num=num//10
print(sum)
```

```python
"""Write a program that checks if a number input by the user is a
 prime number using a for loop."""
n=int(input("Enter the number: "))
for i in range (2,n):
    if(n%i==0):
        print(n," is a not prime number ")
        break
       
else:
    print(n," is a prime number ")
```

```python
n = int(input("Enter the number: "))
if n < 2:
    print(n, "is not prime")
else:
    for i in range(2, n):
        if n % i == 0:
            print(n, "is not prime")
            break
    else:
        print(n, "is prime")
```

```python
"""Write a program that prints the first n Fibonacci numbers,
 where n is input by the user."""
x = int(input("Enter the number: " ))
a=0
b=1
print(a)
print(b)
for i in range (x-2):
    c=a+b
    a=b
    b=c
    print(c)
```
